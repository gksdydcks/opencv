### 14. YOLO (You Only Look Once)
- 딥러닝 활용 객체 탐지


```python
%pip install ultralytics
```

```python
from ultralytics import YOLO
import cv2
```

```python
#YOLO 모델 불러오기
model = YOLO('yolo11n.pt')
#이미지 불러오기
img = cv2.imread("images/bori.jpg")
#신뢰도 0.5로 객체 탐지
results = model.predict(img, conf = 0.4)
#탐지결과를 이미지 위에 그림
annotated_frame = results[0].plot()
#결과 출력
cv2.imshow("img",annotated_frame)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#동영상에 적용
CARS = "videos/cars.mp4"
model = YOLO('yolo11n.pt')
cap = cv2.VideoCapture(CARS)  #변수 선언
# print (cap.isOpened())
while cap.isOpened():
    ret, frame = cap.read()

    if not ret:
        break

    results = model.predict(frame, conf = 0.4)
    annotated_frame = results[0].plot()
    cv2.imshow("video",annotated_frame)

    if cv2.waitKey(1) == ord("q"):
        break
cap.release()
del model  # YOLO 모델 명시적으로 삭제
cv2.destroyAllWindows()
for i in range(5):  # 안정적인 창 닫기 처리
    cv2.waitKey(1)
```

### 실습1. 캠화면에 적용

```python

cap = cv2.VideoCapture(0) 
model = YOLO('yolo11n.pt')
#1
if not cap.isOpened():
    exit()

while True:
    ret, frame = cap.read()
    results = model.predict(frame, conf = 0.4)
    annotated_frame = results[0].plot()

    if not ret:
        break
    cv2.imshow("Camera",annotated_frame)

    if cv2.waitKey(1) == ord("q"):
        break

cap.release()
cv2.destroyAllWindows()
```
